```julia
function logistic_map(x0, r, number = 50, skip = 0) # returns array of {number} elements
    x = x0
    out = []
    if skip != 0
        for i=1:skip
            x = r*x*(1-x)
        end
    end
    for i=1:number
        x = r*x*(1-x)
        append!(out,x)
    end
    return out
end
```

```julia
using Statistics
a1 = logistic_map(0.2,3.72,500000)
a2 = logistic_map(0.200001,3.72,500000)
mean(abs.(a1-a2))
```

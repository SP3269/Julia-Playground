## Bifurcation diagrams using GR

```julia
using Plots; using GR
gr()
```

### Now, to define the map function

```julia
function logistic_map(x0, r, number = 50, skip = 0) # returns array of {number} elements
    x = x0
    out = []
    if skip != 0
        for i=1:skip
            x = r*x*(1-x)
        end
    end
    for i=1:number
        x = r*x*(1-x)
        append!(out,x)
    end
    return out
end
```

### And the plotting function

**Important**: PNG format. The default seems to be SVG, which makes standard hardware run very slowly

```julia
function plot_bfdiagram(rmin = 2.9, rstep = 0.01 , rmax = 3.6; x0 = 0.2, n = 1000, skip = 500, f = logistic_map)
    rvec = []
    xvec = []
    for r in rmin:rstep:rmax
        map = f(x0, r, n, skip)
        append!(xvec,map)
        rtemp = collect(r for i=1:n)
        append!(rvec,rtemp)
    end
    Plots.plot(rvec,xvec,seriestype=:scatter,title="Bifurcation diagram",marker=(:none,1),legend=nothing,fmt=:png)
end
```

```julia
plot_bfdiagram(3.4,0.001,3.55)
```

```julia
plot_bfdiagram(3.564,0.00001,3.565)
```
